## Comparison of models


```python

model = ['Linear Regression', 'SVM', 'XGBoost', 'KNN']
rmse = [17.318, 17.728, 12.782, 22.545]

before_fs = {'model': model, 'rmse': rmse}

df_before_fs = pd.DataFrame(before_fs)

df_before_fs
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_before_fs['model'], df_before_fs['rmse'])
plt.xlabel("Models", fontsize=14)
plt.ylabel("RMSE", fontsize=14)
plt.title("RMSE Comparison of Models Before Feature Selection", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
model = ['Linear Regression', 'SVM', 'XGBoost', 'KNN', 'Gradient Boost', 'Random Forest', 'Ensemble Model']
rmse = [17.264, 17.728, 12.935, 14.840, 19.34, 14.76, 15.86]

after_fs = {'model': model, 'rmse': rmse}

df_after_fs = pd.DataFrame(after_fs)

df_after_fs
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_after_fs['model'], df_after_fs['rmse'])
plt.xlabel("Models", fontsize=14)
plt.ylabel("RMSE", fontsize=14)
plt.title("RMSE Comparison of Models After Feature Selection", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The best model for this particular data set is XGBoost with lowest RMSE value followed by KNN and random forest
